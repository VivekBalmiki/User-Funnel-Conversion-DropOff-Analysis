```python
from datetime import timedelta
import random
import pandas as pd
from faker import Faker

fake = Faker()
data = []

stages = ["Browse", "Add to Cart", "Checkout", "Purchase"]

devices = ["Mobile", "Desktop", "Tablet"]
regions = ["North", "South", "East", "West"]
channels = ["Google Ads", "Organic", "Email", "Social Media"]
categories = ["Electronics", "Fashion", "Home", "Beauty", "Sports"]

for i in range(1, 10001):

    user_id = f"USR{i:05d}"
    user_type = random.choices(["New", "Returning"], weights=[0.7, 0.3])[0]

    sessions = random.randint(1, 3)

    last_session_time = fake.date_time_between(start_date="-30d", end_date="-1d")

    for s in range(sessions):

        session_id = f"{user_id}_SES{s}_{fake.uuid4()[:5]}"

        # Ensure sessions move forward in time
        event_time = last_session_time + timedelta(hours=random.randint(5, 48))
        last_session_time = event_time

        # Session-level fixed attributes
        device = random.choice(devices)
        region = random.choice(regions)
        channel = random.choice(channels)
        category = random.choice(categories)

        # Base probabilities
        probs = {
            "Add to Cart": 0.6,
            "Checkout": 0.45,
            "Purchase": 0.3
        }

        # Behavioral adjustments
        if user_type == "Returning":
            probs["Add to Cart"] += 0.15
            probs["Purchase"] += 0.15

        if device == "Mobile":
            probs["Checkout"] -= 0.05
            probs["Purchase"] -= 0.05

        if channel == "Email":
            probs["Purchase"] += 0.1
        elif channel == "Social Media":
            probs["Add to Cart"] -= 0.05

        if category == "Electronics":
            price = random.uniform(1000, 5000)
            probs["Purchase"] -= 0.1
        else:
            price = random.uniform(200, 2000)

        # Funnel progression
        reached_stages = ["Browse"]

        if random.random() < probs["Add to Cart"]:
            reached_stages.append("Add to Cart")

            if random.random() < probs["Checkout"]:
                reached_stages.append("Checkout")

                # Add slight randomness: some drop after checkout
                if random.random() < probs["Purchase"]:
                    reached_stages.append("Purchase")

        # Generate events
        for stage in reached_stages:

            data.append({
                "User_ID": user_id,
                "Session_ID": session_id,
                "User_Type": user_type,
                "Event": stage,
                "Timestamp": event_time.strftime("%Y-%m-%d %H:%M:%S"),
                "Device": device,
                "Region": region,
                "Channel": channel,
                "Product_Category": category,
                "Revenue": round(price, 2) if stage == "Purchase" else 0
            })

            event_time += timedelta(minutes=random.randint(2, 6))


df = pd.DataFrame(data)

# ✅ Correct bounce: session-level
bounce_sessions = df.groupby("Session_ID")["Event"].nunique() == 1
bounce_sessions = bounce_sessions[bounce_sessions].index

df["Bounce_Flag"] = df["Session_ID"].apply(
    lambda x: "Yes" if x in bounce_sessions else "No"
)

df.to_csv("realistic_funnel_data.csv", index=False)

print("🔥 Realistic funnel dataset generated:", df.shape)
```

```python
import pandas as pd
import numpy as np
import random
from datetime import timedelta
from faker import Faker

fake = Faker()

def generate_realistic_data(n_users=5000):
    data = []
    
    # Configuration
    stages = ["Browse", "Add to Cart", "Checkout", "Purchase"]
    devices = ["Mobile", "Desktop", "Tablet"]
    regions = ["North", "South", "East", "West"]
    channels = ["Google Ads", "Organic", "Email", "Social Media"]
    categories = {
        "Electronics": (1000, 5000),
        "Fashion": (50, 500),
        "Home": (100, 1500),
        "Beauty": (20, 300),
        "Sports": (40, 800)
    }

    for i in range(1, n_users + 1):
        user_id = f"USR{i:05d}"
        user_type = random.choices(["New", "Returning"], weights=[0.7, 0.3])[0]
        
        # A user's journey starts at a random point in the last 30 days
        current_time = fake.date_time_between(start_date="-30d", end_date="-2d")
        
        # Users who don't buy are less likely to keep coming back (Session Decay)
        num_sessions = random.randint(1, 4) if user_type == "Returning" else random.randint(1, 2)

        for s in range(num_sessions):
            session_id = f"{user_id}_S{s}"
            
            # 1. Fix: Ensure sessions are separated by a realistic gap (e.g., 4 to 72 hours)
            current_time += timedelta(hours=random.randint(4, 72))
            
            device = random.choice(devices)
            region = random.choice(regions)
            channel = random.choice(channels)
            category = random.choice(list(categories.keys()))
            min_p, max_p = categories[category]
            session_price = round(random.uniform(min_p, max_p), 2)

            # 2. Fix: Clamped Probabilities (Ensuring they stay between 0 and 1)
            p_cart = 0.4
            p_check = 0.5
            p_buy = 0.4

            if user_type == "Returning": p_cart += 0.2
            if device == "Desktop": p_check += 0.1
            if channel == "Email": p_buy += 0.15
            if category == "Electronics": p_buy -= 0.1  # Higher friction for expensive items

            # Clamp values
            p_cart, p_check, p_buy = [max(0.05, min(0.95, p)) for p in [p_cart, p_check, p_buy]]

            # 3. Funnel Flow
            reached = ["Browse"]
            if random.random() < p_cart:
                reached.append("Add to Cart")
                if random.random() < p_check:
                    reached.append("Checkout")
                    if random.random() < p_buy:
                        reached.append("Purchase")

            # Generate Events with timestamp increments
            event_time = current_time
            for stage in reached:
                data.append({
                    "User_ID": user_id,
                    "Session_ID": session_id,
                    "User_Type": user_type,
                    "Event": stage,
                    "Timestamp": event_time,
                    "Device": device,
                    "Region": region,
                    "Channel": channel,
                    "Product_Category": category,
                    "Revenue": session_price if stage == "Purchase" else 0.0
                })
                # Events within a session happen minutes apart
                event_time += timedelta(minutes=random.randint(1, 8))
            
            # If they purchased, they are less likely to start another session immediately
            if "Purchase" in reached:
                break 

    return pd.DataFrame(data)

# Process and Save
df = generate_realistic_data(10000)

# 4. Fix: Efficient Vectorized Bouncing Logic
# A bounce is a session with only 1 event
session_counts = df.groupby("Session_ID")["Event"].transform("count")
df["Is_Bounce"] = (session_counts == 1)

df.to_csv("clean_funnel_data.csv", index=False)
print(f"✅ Success! Generated {len(df)} rows.")
print(df.head())
```
