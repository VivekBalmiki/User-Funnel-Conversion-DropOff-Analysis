### Funnel Stage Definition and Session-Level Aggregation

```python
# Define funnel stages
funnel_stages = ['Browse', 'Add to cart', 'Checkout', 'Purchase']

# Precompute stage ranking (avoids rebuilding dict every function call)
stage_rank = {stage: i for i, stage in enumerate(funnel_stages)}

# Sort data BEFORE grouping (important for correct event order)
df = df.sort_values(['Session_ID', 'Timestamp'])

# Create session-level summary
session_summary = (
    df.groupby('Session_ID')
      .agg(
          User_ID=('User_ID', 'first'),
          Session_Start=('Timestamp', 'min'),
          Session_End=('Timestamp', 'max'),
          Event_Sequence=('Event', list),
          Device=('Device', 'first'),
          Region=('Region', 'first'),
          Channel=('Channel', 'first'),
          Product_Category=('Product_Category', 'first'),
          Revenue=('Revenue', 'first'),
          Bounce_Flag=('Bounce_Flag', 'first')
      )
      .reset_index()
)

# Session duration (vectorized)
session_summary['Session_Duration_Min'] = (
    (session_summary['Session_End'] - session_summary['Session_Start'])
    .dt.total_seconds()
    .div(60)
)

# Optimized funnel stage detection
def get_max_funnel_stage(events):
    max_stage = -1
    for e in events:
        rank = stage_rank.get(e, -1)
        if rank > max_stage:
            max_stage = rank
    return funnel_stages[max_stage] if max_stage >= 0 else funnel_stages[0]

# Apply function
session_summary['Max_Funnel_Stage'] = session_summary['Event_Sequence'].apply(get_max_funnel_stage)

print("Session Summary Created:")
display(session_summary.head())
```

### Overall Funnel Analysis

```python
# Calculate overall funnel metrics
funnel_metrics = []

total_sessions = len(session_summary)

for i, stage in enumerate(funnel_stages):
    if i == 0:
        # Browse stage: all sessions
        count = total_sessions
    else:
        # Other stages: sessions that reached at least this stage
        count = session_summary['Max_Funnel_Stage'].apply(
            lambda x: funnel_stages.index(x) >= i
        ).sum()
        
    funnel_metrics.append({
        'Stage': stage,
        'Sessions': count,
        'Stage_Order': i
    })

# Convert to DataFrame
funnel_df = pd.DataFrame(funnel_metrics)

# Conversion rate from first stage
funnel_df['Conversion_Rate'] = (funnel_df['Sessions'] / total_sessions * 100).round(2)

# Drop-off rate from previous stage
funnel_df['Drop_Off_Rate'] = (1 - funnel_df['Sessions'] / funnel_df['Sessions'].shift(1)).fillna(0) * 100
funnel_df['Drop_Off_Rate'] = funnel_df['Drop_Off_Rate'].round(2)

print("🎯 Overall Funnel Analysis:")
display(funnel_df)
```

### Visualization - Overall Funnel

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Prepare subplot
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Funnel Conversion Rates', 
        'Stage-to-Stage Drop-off (%)',
        'Revenue by Funnel Stage', 
        'Avg Session Duration by Stage (min)'
    ),
    specs=[[{"secondary_y": False}, {"secondary_y": False}],
           [{"secondary_y": False}, {"secondary_y": False}]]
)

# 1️⃣ Funnel Conversion Rates
fig.add_trace(
    go.Bar(
        x=funnel_df['Stage'], 
        y=funnel_df['Sessions'],
        text=funnel_df['Sessions'],
        textposition='auto',
        name='Sessions',
        marker_color='lightblue'
    ),
    row=1, col=1
)

# 2️⃣ Stage-to-Stage Drop-off
fig.add_trace(
    go.Scatter(
        x=funnel_df['Stage'], 
        y=funnel_df['Drop_Off_Rate'],
        mode='lines+markers+text',
        text=funnel_df['Drop_Off_Rate'].apply(lambda x: f"{x:.2f}%"),
        textposition='top center',
        name='Drop-off Rate',
        line=dict(color='red', width=3)
    ),
    row=1, col=2
)

# 3️⃣ Revenue by Funnel Stage
revenue_by_stage = session_summary.groupby('Max_Funnel_Stage')['Revenue'].sum().reindex(funnel_stages).fillna(0)
fig.add_trace(
    go.Bar(
        x=revenue_by_stage.index, 
        y=revenue_by_stage.values,
        text=[f"${x:,.0f}" for x in revenue_by_stage.values],
        textposition='auto',
        name='Revenue',
        marker_color='green'
    ),
    row=2, col=1
)

# 4️⃣ Avg Session Duration by Stage
duration_by_stage = session_summary.groupby('Max_Funnel_Stage')['Session_Duration_Min'].mean().reindex(funnel_stages).fillna(0)
fig.add_trace(
    go.Bar(
        x=duration_by_stage.index, 
        y=duration_by_stage.values.round(2),
        text=duration_by_stage.values.round(2),
        textposition='auto',
        name='Avg Session Duration',
        marker_color='orange'
    ),
    row=2, col=2
)

# Update layout
fig.update_layout(
    height=800, width=1000,
    title_text="🎯 Comprehensive Funnel Analysis Dashboard",
    showlegend=False,
    template='plotly_white'
)

fig.show()
```

### Channel Performance Analysis

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Standardize stage names
session_summary['Max_Funnel_Stage'] = session_summary['Max_Funnel_Stage'].str.strip().str.title()

channel_funnel = []

for channel, group in session_summary.groupby('Channel'):
    total_sessions = len(group)
    metrics = {'Channel': channel, 'Total_Sessions': total_sessions}

    # Stage-level sessions & rates
    for i, stage in enumerate(funnel_stages):
        if i == 0:
            count = total_sessions
        else:
            count = (group['Max_Funnel_Stage'].apply(lambda x: funnel_stages.index(x) >= i)).sum()
        metrics[f'{stage}_Sessions'] = count
        metrics[f'{stage}_Rate'] = round(count / total_sessions * 100, 2)

    # Revenue metrics
    purchase_sessions = group[group['Max_Funnel_Stage'] == 'Purchase']
    metrics['Total_Revenue'] = purchase_sessions['Revenue'].sum()
    metrics['AOV'] = purchase_sessions['Revenue'].mean() if len(purchase_sessions) > 0 else 0
    metrics['Conversion_Rate'] = round(len(purchase_sessions) / total_sessions * 100, 2)

    channel_funnel.append(metrics)

channel_df = pd.DataFrame(channel_funnel).sort_values('Conversion_Rate', ascending=False)

print("📊 Channel Performance Analysis:")
display(channel_df.round(2))

# --- Interactive Visualizations using Plotly ---
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Conversion Rate by Channel', 'Total Revenue by Channel',
                    'Average Order Value (AOV)', 'Session Distribution'),
    specs=[[{"type": "bar"}, {"type": "bar"}],
           [{"type": "bar"}, {"type": "pie"}]]
)

# Conversion Rate
fig.add_trace(go.Bar(
    x=channel_df['Channel'], y=channel_df['Conversion_Rate'],
    text=channel_df['Conversion_Rate'].apply(lambda x: f"{x}%"),
    textposition='auto', name='Conversion Rate', marker_color='lightblue'
), row=1, col=1)

# Total Revenue
fig.add_trace(go.Bar(
    x=channel_df['Channel'], y=channel_df['Total_Revenue'],
    text=channel_df['Total_Revenue'].apply(lambda x: f"${x:,.0f}"),
    textposition='auto', name='Total Revenue', marker_color='green'
), row=1, col=2)

# Average Order Value
fig.add_trace(go.Bar(
    x=channel_df['Channel'], y=channel_df['AOV'],
    text=channel_df['AOV'].apply(lambda x: f"${x:,.2f}"),
    textposition='auto', name='AOV', marker_color='orange'
), row=2, col=1)

# Session Distribution
fig.add_trace(go.Pie(
    labels=channel_df['Channel'],
    values=channel_df['Total_Sessions'],
    textinfo='label+percent',
    name='Session Distribution'
), row=2, col=2)

fig.update_layout(height=800, width=1000, title_text="📊 Channel Performance Dashboard", showlegend=False)
fig.show()
```

### Regional Analysis

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# --- Regional Performance Summary ---
regional_analysis = session_summary.groupby('Region').agg(
    Total_Sessions=('Session_ID', 'count'),
    Total_Revenue=('Revenue', 'sum'),
    Avg_Session_Duration=('Session_Duration_Min', 'mean')
).round(2)

# Converted sessions & Conversion Rate
converted_sessions = session_summary[session_summary['Max_Funnel_Stage'] == 'Purchase'].groupby('Region').size()
regional_analysis['Converted_Sessions'] = converted_sessions
regional_analysis['Converted_Sessions'] = regional_analysis['Converted_Sessions'].fillna(0).astype(int)

regional_analysis['Conversion_Rate'] = (
    regional_analysis['Converted_Sessions'] / regional_analysis['Total_Sessions'] * 100
).round(2)

regional_analysis['AOV'] = (
    regional_analysis['Total_Revenue'] / regional_analysis['Converted_Sessions']
).fillna(0).round(2)

print("🌎 Regional Performance Summary:")
display(regional_analysis)

# --- Regional Funnel Progression (Interactive) ---
funnel_records = []

for region, group in session_summary.groupby('Region'):
    total_sessions = len(group)
    for i, stage in enumerate(funnel_stages):
        if i == 0:
            count = total_sessions
        else:
            count = (group['Max_Funnel_Stage'].apply(lambda x: funnel_stages.index(x) >= i)).sum()
        funnel_records.append({'Region': region, 'Stage': stage, 'Sessions': count})

regional_funnel_df = pd.DataFrame(funnel_records)

# Interactive line chart using Plotly
fig = go.Figure()

for region in regional_funnel_df['Region'].unique():
    df_region = regional_funnel_df[regional_funnel_df['Region'] == region]
    fig.add_trace(go.Scatter(
        x=df_region['Stage'], y=df_region['Sessions'],
        mode='lines+markers+text',
        text=df_region['Sessions'],
        textposition='top center',
        name=region
    ))

fig.update_layout(
    title="🌎 Funnel Progression by Region",
    xaxis_title="Funnel Stage",
    yaxis_title="Sessions",
    template='plotly_white',
    height=500,
    width=900
)

fig.show()
```

### Device and Product Category Analysis

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# --- Device Performance ---
device_analysis = session_summary.groupby('Device').agg(
    Total_Sessions=('Session_ID', 'count'),
    Revenue=('Revenue', 'sum'),
    Avg_Session_Duration=('Session_Duration_Min', 'mean'),
    Purchases=('Max_Funnel_Stage', lambda x: (x == 'Purchase').sum())
).round(2)

device_analysis['Conversion_Rate'] = (device_analysis['Purchases'] / device_analysis['Total_Sessions'] * 100).round(2)
device_analysis['AOV'] = (device_analysis['Revenue'] / device_analysis['Purchases']).fillna(0).round(2)

print("📱 Device Performance:")
display(device_analysis)

# --- Product Category Performance ---
product_analysis = session_summary.groupby('Product_Category').agg(
    Total_Sessions=('Session_ID', 'count'),
    Revenue=('Revenue', 'sum'),
    Purchases=('Max_Funnel_Stage', lambda x: (x == 'Purchase').sum())
).round(2)

product_analysis['Conversion_Rate'] = (product_analysis['Purchases'] / product_analysis['Total_Sessions'] * 100).round(2)
product_analysis['AOV'] = (product_analysis['Revenue'] / product_analysis['Purchases']).fillna(0).round(2)
product_analysis['Revenue_Per_Session'] = (product_analysis['Revenue'] / product_analysis['Total_Sessions']).round(2)

print("🏷️ Product Category Performance:")
display(product_analysis)

# --- Interactive Combined Visualization ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Conversion Rate by Device Type', 'Conversion Rate by Product Category')
)

# Device conversion rates
fig.add_trace(go.Bar(
    x=device_analysis.index,
    y=device_analysis['Conversion_Rate'],
    text=device_analysis['Conversion_Rate'].apply(lambda x: f"{x:.2f}%"),
    textposition='auto',
    name='Device Conversion Rate',
    marker_color='skyblue'
), row=1, col=1)

# Product category conversion rates
fig.add_trace(go.Bar(
    x=product_analysis.index,
    y=product_analysis['Conversion_Rate'],
    text=product_analysis['Conversion_Rate'].apply(lambda x: f"{x:.2f}%"),
    textposition='auto',
    name='Product Category Conversion Rate',
    marker_color='orange'
), row=1, col=2)

fig.update_layout(
    height=500, width=1000,
    title_text="📊 Device & Product Category Performance",
    showlegend=False,
    template='plotly_white'
)

fig.show()
```

### Time - Based Analysis 

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# --- Daily Metrics ---
daily_metrics = df.groupby('Date').agg(
    Daily_Sessions=('Session_ID', 'nunique'),
    Daily_Users=('User_ID', 'nunique'),
    Daily_Revenue=('Revenue', 'sum')
)

# Daily conversions
daily_conversions = session_summary[session_summary['Max_Funnel_Stage'] == 'Purchase'].groupby(
    session_summary['Session_Start'].dt.date
).size()

daily_metrics['Daily_Conversions'] = daily_conversions
daily_metrics['Daily_Conversions'] = daily_metrics['Daily_Conversions'].fillna(0).astype(int)
daily_metrics['Daily_Conversion_Rate'] = (daily_metrics['Daily_Conversions'] / daily_metrics['Daily_Sessions'] * 100).round(2)

print("📅 Daily Performance Trends:")
display(daily_metrics.tail(10))

# --- Hourly Metrics ---
hourly_sessions = df.groupby('Hour').agg(
    Hourly_Sessions=('Session_ID', 'nunique'),
    Hourly_Revenue=('Revenue', 'sum')
)

hourly_conversions = session_summary[session_summary['Max_Funnel_Stage'] == 'Purchase'].copy()
hourly_conversions['Hour'] = hourly_conversions['Session_Start'].dt.hour
hourly_conversion_counts = hourly_conversions.groupby('Hour').size()

hourly_sessions['Hourly_Conversions'] = hourly_conversion_counts
hourly_sessions['Hourly_Conversions'] = hourly_sessions['Hourly_Conversions'].fillna(0).astype(int)
hourly_sessions['Hourly_Conversion_Rate'] = (hourly_sessions['Hourly_Conversions'] / hourly_sessions['Hourly_Sessions'] * 100).round(2)

# --- Interactive Visualization ---
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Daily Sessions Over Time',
        'Daily Conversion Rate (%)',
        'Hourly Sessions by Hour of Day',
        'Hourly Conversion Rate (%)'
    ),
    specs=[[{"type": "scatter"}, {"type": "scatter"}],
           [{"type": "bar"}, {"type": "bar"}]]
)

# Daily Sessions
fig.add_trace(go.Scatter(
    x=daily_metrics.index, y=daily_metrics['Daily_Sessions'],
    mode='lines+markers', name='Daily Sessions', line=dict(color='blue')
), row=1, col=1)

# Daily Conversion Rate
fig.add_trace(go.Scatter(
    x=daily_metrics.index, y=daily_metrics['Daily_Conversion_Rate'],
    mode='lines+markers', name='Daily Conversion Rate', line=dict(color='green')
), row=1, col=2)

# Hourly Sessions
fig.add_trace(go.Bar(
    x=hourly_sessions.index, y=hourly_sessions['Hourly_Sessions'],
    name='Hourly Sessions', marker_color='orange'
), row=2, col=1)

# Hourly Conversion Rate
fig.add_trace(go.Bar(
    x=hourly_sessions.index, y=hourly_sessions['Hourly_Conversion_Rate'],
    name='Hourly Conversion Rate', marker_color='red'
), row=2, col=2)

fig.update_layout(
    height=700, width=1000,
    title_text="📊 Daily & Hourly Performance Trends",
    showlegend=False,
    template='plotly_white'
)

fig.show()
```
